```python
green_df = green_df.rename(columns={
    'lb_code': 'Borough',
    'percent_green':'Percent Green Areas'
})

green_df['Borough'] = green_df['Borough'].replace({
    'City of Westminster': 'Westminster'
})
```

```python
green_df= green_df[['Borough', 'Percent Green Areas']].copy()
green_df
```

### Merging

```python
merged_df = pd.merge(mental_df, green_df, on = 'Borough', how='inner')

merged_df = merged_df[merged_df['Borough'] !='City of London']

merged_df.head(5)
```

### Visualizing

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=merged_df, x='Percent Green Areas', y='Life Satisfaction')

plt.title('Relationship Between Green Space And Life Satisfaction')
plt.xlabel('Percent Green Areas %')
plt.ylabel('Life Satisfaction 1-10')
plt.grid(True)
plt.show()
          
```

```python
import numpy as np
```

```python
merged_sorted = merged_df.sort_values(by='Life Satisfaction', ascending = False)

x = np.arange(len(merged_sorted))
width = 0.4

plt.figure(figsize=(14,7))
plt.bar(x - width/2, merged_sorted['Life Satisfaction'], width, label='Life Satisfaction', color='blue')
plt.bar(x + width/2, merged_sorted['Percent Green Areas'], width, label='Percent Green Areas', color='green')

plt.xticks(x, merged_sorted['Borough'], rotation=90)
plt.title('Relationship Between Green Space And Life Satisfaction By Borough')
plt.ylabel('Score/Percent')
plt.legend()
plt.show()
```

```python
merged_sorted = merged_df.sort_values(by='Percent Green Areas', ascending = False)

fig, ax1 = plt.subplots(figsize=(14,7))

ax1.bar(merged_sorted['Borough'], merged_sorted['Percent Green Areas'], color = 'green')
ax1.set_ylabel('Percent Green Areas %', color='green')
ax1.tick_params(axis='y', labelcolor = 'green')
ax1.set_xticks(range(len(merged_sorted)))
ax1.set_xticklabels(merged_sorted['Borough'], rotation=90)

ax2 = ax1.twinx()
ax2.plot(merged_sorted['Borough'], merged_sorted['Life Satisfaction'], color = 'blue')
ax2.set_ylabel('Life Satisfaction Score', color='blue')
ax2.tick_params(axis='y', labelcolor = 'blue')

#min_val = merged_sorted['Life Satisfaction'].min()
#max_val = merged_sorted['Life Satisfaction'].max()
ax2.set_ylim(7,8)

plt.title('Relationship Between Green Space And Life Satisfaction By Borough')
plt.show()
```
