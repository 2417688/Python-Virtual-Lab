### Transforming Data

```python
df = pd.read_csv('data/loan_data.csv')
df
```

```python
df['Net Assets'] = df['Balance'] - df['Debt']
```

```python
df
```

```python
df.replace({'Short Term': "6 Month Loan", 'Long Term': "12 Month Loan"})
```

```python
df.replace({'Short': "6 Month", 'Long': "12 Month"}, regex=True)
```

```python
def debtclassify(debt):
    if debt > 1000:
        return "High"
    else:
        return "Low"
```

```python
df['Debt'].map(debtclassify)
```

```python
df['Debt'].map(lambda debt:'High' if debt > 1000 else 'Low')
```
