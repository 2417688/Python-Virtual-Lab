### Aggregating

```python
df[['Term', 'Balance']].groupby('Term').sum()
```

```python
df[['Term', 'Balance']].groupby('Term').mean()
```

```python
df[['Term','Default', 'Balance']].groupby(['Term', 'Default']).mean()
```

```python
df[['Debt', 'Term']].groupby('Term').transform('mean')
```

```python
df['MeanTermDebt'] = df[['Debt', 'Term']].groupby('Term').transform('mean')
```
